```python
from ultralytics import YOLO
import os
import cv2

model = YOLO('yolov8n.pt')  

image_dir = r"C:\Users\sunil\Downloads\cogniable\frames\val"
label_dir = r"C:\Users\sunil\Downloads\cogniable\labels\val"
os.makedirs(label_dir, exist_ok=True)

PERSON_CLASS_ID = 0

for image_file in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_file)
    image = cv2.imread(image_path)

    results = model(image)

    person_detected = False

    label_filename = os.path.splitext(image_file)[0] + '.txt'
    label_path = os.path.join(label_dir, label_filename)
    
    with open(label_path, 'w') as f:
        for result in results:
            for bbox, conf, cls in zip(result.boxes.xyxy, result.boxes.conf, result.boxes.cls):
                if int(cls) == PERSON_CLASS_ID:  # Filter for "person" class only
                    person_detected = True
                    x_center = (bbox[0] + bbox[2]) / 2 / image.shape[1]
                    y_center = (bbox[1] + bbox[3]) / 2 / image.shape[0]
                    width = (bbox[2] - bbox[0]) / image.shape[1]
                    height = (bbox[3] - bbox[1]) / image.shape[0]
                    f.write(f'{int(cls)} {x_center} {y_center} {width} {height}\n')

    if not person_detected:
        os.remove(image_path)
        os.remove(label_path)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8s.pt') 
model.train(data=r'C:\Users\sunil\Downloads\cogniable\dataset.yaml',
            epochs=5,
            imgsz=640,
            batch=16,
            name='yolov8_person_detection',
            device='cpu',
            verbose=True) 
```

```python
import cv2
import torch
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

model = YOLO(r"C:\Users\sunil\Downloads\cogniable.pt") 
tracker = DeepSort(max_age=30, nn_budget=70)

cap = cv2.VideoCapture(0) 

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('webcam_output_with_detections.mp4', fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))

try:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame)

        detections = []
        for result in results:
            for bbox, conf, cls in zip(result.boxes.xyxy, result.boxes.conf, result.boxes.cls):
                if cls.item() == 0 and conf > 0.5:  # Only process 'person' class (class 0)
                    x1, y1, x2, y2 = bbox.tolist()  # Convert tensor to list of floats
                    detections.append(([x1, y1, x2, y2], conf.item(), cls.item()))

        tracks = tracker.update_tracks(detections, frame=frame)

        for track in tracks:
            if not track.is_confirmed():
                continue
            track_id = track.track_id
            ltrb = track.to_ltrb()
            x1, y1, x2, y2 = map(int, ltrb)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f'ID: {track_id}', (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        out.write(frame)

except KeyboardInterrupt:
    print("Detection stopped by user.")

finally:
    cap.release()
    out.release()
    print("Video saved and resources released.")
```
